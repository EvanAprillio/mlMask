```python
# Import libraries
import pickle
import tensorflow as tf
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
from PIL import Image
     
```

```python
model2 = load_model('model_seq.h5')
```

```python
img=image.load_img("predict.jpeg")
img
```

```python
img = image.img_to_array(img) / 220
img = tf.image.resize(img, (220,220))
img = tf.expand_dims(img, axis=0)
```

```python
prediction = int(tf.round(model2.predict(x = img)).numpy()[0][0])

def classpredict(value):
    if value == 0:
        return "WithMask"
    elif value == 1:
        return "WithoutMask"

print('Picture of someone  ', classpredict(prediction))
```
